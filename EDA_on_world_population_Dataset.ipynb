```python
numeric_df = df.select_dtypes(include='number')  # Select only numeric columns

plt.figure(figsize=(20, 7))  # Set the figure size
sns.heatmap(numeric_df.corr(), annot=True)  # Create the heatmap with annotations

plt.show()
```

```python
# Exclude 'Continent' column from the calculation
numeric_columns = df.select_dtypes(include=['number']).columns
numeric_df = df[numeric_columns]

# Calculate mean for numeric columns grouped by 'Continent'
continent_mean = numeric_df.groupby(df['Continent'])[['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']].mean().sort_values(by='2022 Population', ascending=False)

continent_mean
```

```python
# Reverse the list of column names
reversed_columns = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population'][::-1]

# Group by 'Continent' and calculate the mean for the reversed columns
df2 = numeric_df.groupby(df['Continent'])[reversed_columns].mean().sort_values(by='2022 Population', ascending=False)

df2
```

```python
df3 = df2.transpose()
df3
```

```python
df.boxplot() #looking at outliers
```

```python
# Create histograms for each population year
plt.figure(figsize=(20, 7))
for column in reversed_columns:
    plt.hist(df[column], bins=20, alpha=0.5, label=column, density=True, orientation='horizontal')

plt.xlabel('Population')
plt.ylabel('Density')
plt.title('Population Distribution Over Time')

plt.legend()
plt.show()
```

```python
# Create scatter plots for each pair of population years
plt.figure(figsize=(20, 8))
for i in range(len(reversed_columns)):
    for j in range(i+1, len(reversed_columns)):
        plt.scatter(df[reversed_columns[i]], df[reversed_columns[j]], alpha=0.5, s=50, label=f'{reversed_columns[i]} vs {reversed_columns[j]}')

plt.title('Population Scatter Plot')
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = df[reversed_columns].corr()

# Create heatmap
plt.figure(figsize=(20, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap of Population Years')
plt.show()
```

```python
# Create box plots for each population year
plt.figure(figsize=(20, 8))
df[reversed_columns].boxplot()
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Distribution Over Time')
plt.show()
```

```python
# Create bar plots for each continent
plt.figure(figsize=(10, 8))
df2.plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.title('Population by Continent')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Continent', hue='Continent', data=df, palette='viridis', legend=False)
plt.title('Distribution of Countries by Continent')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Detect outliers using z-score method
from scipy.stats import zscore
z_scores = zscore(df[reversed_columns])
outliers = (z_scores > 3).sum(axis=0)

# Remove outliers from dataframe
df_cleaned = df[(z_scores < 3).all(axis=1)]
```

```python
#after removing outliers
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_cleaned[reversed_columns])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Distribution Over Time (After Outlier Removal)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair plot for numeric columns (not necessary takes a lot of time)
sns.pairplot(df[reversed_columns], diag_kind='kde')
plt.suptitle('Pair Plot of Population Years', y=1.02)
plt.show()
```
